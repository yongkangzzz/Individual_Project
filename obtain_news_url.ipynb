```python
from newspaper import Article
import requests
from bs4 import BeautifulSoup as bs
def get_title(url):
    news = Article(url)
    news.download()
    news.parse()
    title = news.title
    return title
def get_text(url):
    news = Article(url)
    news.download()
    news.parse()
    title = news.title
    if url.startswith('https://www.bbc'):
        article = requests.get(url)
        soup = bs(article.content, "html.parser")
        text = ''
        for EachPart in soup.select('div[class*="RichTextContainer"]'):
            for p in EachPart.find_all("p"):
                text = text+' '+p.text
        if len(text)==0:
            text=news.text
    else :
        text = news.text
    text = title+text
    if len(text)>1000:
      text = text[0:1000]
    return text
```

```python
df = pd.read_csv('/content/drive/My Drive/mycolab/ob/trainset11.csv')
titles=[]
postitles=[]
negtitles=[]
for i in range(df.shape[0]):
  title = df['title'].loc[i]
  url = df['url'].loc[i]
  sentiment =df['label'].loc[i]
  if type(title) != float and url not in other:
    title = title.strip("\n").strip("\t")
    titles.append(title)
    if sentiment==1:
      postitles.append(title)
    elif sentiment==0:
      negtitles.append(title)
```

```python
import re
from collections import Counter
cnt = Counter()
for title in titles:
    title = title.lower()
    str = ''.join(re.findall('[a-z ]',title))
    alist=str.split()

    for item in alist:
        if item.isalpha():
            cnt.update({item:1})

poscnt = Counter()
for title in postitles:
    title = title.lower()
    str = ''.join(re.findall('[a-z ]',title))
    alist=str.split()

    for item in alist:
        if item.isalpha():
            poscnt.update({item:1})

negcnt = Counter()
for title in negtitles:
    title = title.lower()
    str = ''.join(re.findall('[a-z ]',title))
    alist=str.split()

    for item in alist:
        if item.isalpha():
            negcnt.update({item:1})
```
